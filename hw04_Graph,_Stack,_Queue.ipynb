#232. Implement Queue using Stacks
https://leetcode.com/problems/implement-queue-using-stacks/description/


Implement a first in first out (FIFO) queue using only two stacks. The implemented queue should support all the functions of a normal queue (push, peek, pop, and empty).

Implement the MyQueue class:

void push(int x) Pushes element x to the back of the queue.
int pop() Removes the element from the front of the queue and returns it.
int peek() Returns the element at the front of the queue.
boolean empty() Returns true if the queue is empty, false otherwise.
Notes:

You must use only standard operations of a stack, which means only push to top, peek/pop from top, size, and is empty operations are valid.
Depending on your language, the stack may not be supported natively. You may simulate a stack using a list or deque (double-ended queue) as long as you use only a stack's standard operations.


Example 1:

Input
["MyQueue", "push", "push", "peek", "pop", "empty"]
[[], [1], [2], [], [], []]
Output
[null, null, null, 1, 1, false]

Explanation
MyQueue myQueue = new MyQueue();
myQueue.push(1); // queue is: [1]
myQueue.push(2); // queue is: [1, 2] (leftmost is front of the queue)
myQueue.peek(); // return 1
myQueue.pop(); // return 1, queue is [2]
myQueue.empty(); // return false


Constraints:

1 <= x <= 9
At most 100 calls will be made to push, pop, peek, and empty.
All the calls to pop and peek are valid.


Follow-up: Can you implement the queue such that each operation is amortized O(1) time complexity? In other words, performing n operations will take overall O(n) time even if one of those operations may take longer.


🧾 문제 설명: Implement Queue using Stacks
✅ 목표
큐 (FIFO: First-In-First-Out) 구조를 스택 (LIFO: Last-In-First-Out) 2개로 구현

지원해야 하는 연산:

push(x): 큐 뒤에 원소 x 추가

pop(): 큐 앞의 원소 제거 후 반환

peek(): 큐 앞의 원소 반환 (제거 X)

empty(): 큐가 비었는지 여부 확인



```python
class MyQueue:
    def __init__(self):
        self.stk1 = []  # 입력용 스택
        self.stk2 = []  # 출력용 스택

    def push(self, x: int) -> None:
        # 항상 stk1에만 push
        self.stk1.append(x)

    def pop(self) -> int:
        # 큐의 앞쪽 원소를 꺼내야 하므로, 필요 시 stk1 → stk2 이동
        self.move()
        return self.stk2.pop()

    def peek(self) -> int:
        # 큐의 앞쪽 값을 보기만 함 (pop은 안 함)
        self.move()
        return self.stk2[-1]

    def empty(self) -> bool:
        # 두 스택 모두 비어 있어야 큐가 빈 상태
        return not self.stk1 and not self.stk2

    def move(self):
        # stk2가 비어 있으면, stk1의 데이터를 stk2로 역전 이동
        if not self.stk2:
            while self.stk1:
                self.stk2.append(self.stk1.pop())
```

#394. Decode String
https://leetcode.com/problems/decode-string/description/

Given an encoded string, return its decoded string.

The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2[4].

The test cases are generated so that the length of the output will never exceed 105.



Example 1:

Input: s = "3[a]2[bc]"
Output: "aaabcbc"
Example 2:

Input: s = "3[a2[c]]"
Output: "accaccacc"
Example 3:

Input: s = "2[abc]3[cd]ef"
Output: "abcabccdcdcdef"


Constraints:

1 <= s.length <= 30
s consists of lowercase English letters, digits, and square brackets '[]'.
s is guaranteed to be a valid input.
All the integers in s are in the range [1, 300].


🧾 문제 설명: Decode String
✅ 문제 요약
입력: 인코딩된 문자열 s

출력: 디코딩된 원래 문자열

✅ 인코딩 규칙
k[encoded_string] → encoded_string을 k번 반복한 문자열

중첩도 가능: "3[a2[c]]" → "accaccacc"



```python
class Solution:
    def decodeString(self, s: str) -> str:
        s1, s2 = [], []      # s1: 숫자 스택, s2: 문자열 스택
        num, res = 0, ''     # 현재 숫자 누적용, 현재 문자열 누적용

        for c in s:
            if c.isdigit():
                # 숫자가 여러 자리일 수 있으므로 누적
                num = num * 10 + int(c)

            elif c == '[':
                # 현재 숫자와 문자열 상태를 스택에 저장하고 초기화
                s1.append(num)
                s2.append(res)
                num, res = 0, ''

            elif c == ']':
                # 스택에서 숫자와 이전 문자열을 꺼내 반복 적용
                repeat_count = s1.pop()
                prev_str = s2.pop()
                res = prev_str + res * repeat_count

            else:
                # 일반 문자는 결과 문자열에 추가
                res += c

        return res  # 최종 결과 반환
```

#2327. Number of People Aware of a Secret
https://leetcode.com/problems/number-of-people-aware-of-a-secret/description/


On day 1, one person discovers a secret.

You are given an integer delay, which means that each person will share the secret with a new person every day, starting from delay days after discovering the secret. You are also given an integer forget, which means that each person will forget the secret forget days after discovering it. A person cannot share the secret on the same day they forgot it, or on any day afterwards.

Given an integer n, return the number of people who know the secret at the end of day n. Since the answer may be very large, return it modulo 109 + 7.



Example 1:

Input: n = 6, delay = 2, forget = 4
Output: 5
Explanation:
Day 1: Suppose the first person is named A. (1 person)
Day 2: A is the only person who knows the secret. (1 person)
Day 3: A shares the secret with a new person, B. (2 people)
Day 4: A shares the secret with a new person, C. (3 people)
Day 5: A forgets the secret, and B shares the secret with a new person, D. (3 people)
Day 6: B shares the secret with E, and C shares the secret with F. (5 people)
Example 2:

Input: n = 4, delay = 1, forget = 3
Output: 6
Explanation:
Day 1: The first person is named A. (1 person)
Day 2: A shares the secret with B. (2 people)
Day 3: A and B share the secret with 2 new people, C and D. (4 people)
Day 4: A forgets the secret. B, C, and D share the secret with 3 new people. (6 people)


Constraints:

2 <= n <= 1000
1 <= delay < forget <= n



🧾 문제 설명: People Aware of a Secret
✅ 문제 요약
입력:

n: 총 일수 (1 ≤ n ≤ 1000)

delay: 비밀을 알게 된 사람이 몇 일이 지나야 다른 사람에게 공유할 수 있는지

forget: 비밀을 몇 일이 지나면 까먹는지 (까먹은 날에는 공유 불가능)

출력:

n일이 지난 후 비밀을 알고 있는 사람의 수 (mod 1e9+7)

```python
class Solution:
    def peopleAwareOfSecret(self, n: int, delay: int, forget: int) -> int:
        m = (n << 1) + 10         # 넉넉한 배열 크기
        d = [0] * m               # 각 날짜마다 알고 있는 사람 변화량 (prefix sum용)
        cnt = [0] * m             # 각 날짜에 새로 알게 된 사람 수
        cnt[1] = 1                # Day 1: 최초 1명

        for i in range(1, n + 1):
            if cnt[i]:
                d[i] += cnt[i]                  # 그날 알고 있는 사람 수 추가
                d[i + forget] -= cnt[i]         # forget일에 비밀을 까먹으므로 감소
                nxt = i + delay                 # delay 후부터 공유 시작
                while nxt < i + forget:         # forget 전까지 공유 가능
                    cnt[nxt] += cnt[i]          # 새로 알게 되는 사람 추가
                    nxt += 1

        mod = 10**9 + 7
        return sum(d[:n + 1]) % mod             # Day 1 ~ n까지 알고 있는 사람 합계
```
