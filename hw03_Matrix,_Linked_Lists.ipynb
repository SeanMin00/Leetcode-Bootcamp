#73. Palindrome Linked List.
https://leetcode.com/problems/set-matrix-zeroes/description/


image.png


```python
from typing import List

class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        m, n = len(matrix), len(matrix[0])  # m:  , n:  

        # 1. 0      
        row = [False] * m  #  
        col = [False] * n  #  

        # 2.   0     True 
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    row[i] = True
                    col[j] = True

        # 3.   ,    or    0 
        for i in range(m):
            for j in range(n):
                if row[i] or col[j]:
                    matrix[i][j] = 0
```

#234. Palindrome Linked List
https://leetcode.com/problems/palindrome-linked-list/description/

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        # 1. fast slow    
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # 2.     (slow )
        pre, cur = None, slow.next
        while cur:
            t = cur.next       #   
            cur.next = pre     #       ()
            pre, cur = cur, t  # pre cur   

        # 3.  (head)   (pre) 
        while pre:
            if pre.val != head.val:
                return False   #    
            pre = pre.next
            head = head.next

        # 4.   
        return True
```
