#73. Palindrome Linked List.
https://leetcode.com/problems/set-matrix-zeroes/description/


image.png


```python
from typing import List

class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        m, n = len(matrix), len(matrix[0])  # m: 행 개수, n: 열 개수

        # 1. 0이 있는 행과 열을 기록할 배열 초기화
        row = [False] * m  # 행 마크용
        col = [False] * n  # 열 마크용

        # 2. 행렬을 순회하면서 0이 발견되면 해당 행과 열에 True 마크
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    row[i] = True
                    col[j] = True

        # 3. 다시 행렬을 순회하면서, 해당 위치의 행 or 열이 마크되어 있으면 0으로 변경
        for i in range(m):
            for j in range(n):
                if row[i] or col[j]:
                    matrix[i][j] = 0
```

#234. Palindrome Linked List
https://leetcode.com/problems/palindrome-linked-list/description/


문제 설명
✅ 문제 요약
입력: 단일 연결 리스트 head

목표: 연결 리스트가 **회문(palindrome)**인지 확인

회문: 앞에서 읽어도, 뒤에서 읽어도 같은 값을 가진 구조

출력: True 또는 False

image.png

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        # 1. fast와 slow를 사용해서 중간 지점 찾기
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # 2. 리스트 후반부 반전 시작 (slow는 중간)
        pre, cur = None, slow.next
        while cur:
            t = cur.next       # 다음 노드를 저장
            cur.next = pre     # 현재 노드가 이전 노드를 가리키게 함 (역방향)
            pre, cur = cur, t  # pre와 cur을 한 칸씩 이동

        # 3. 앞 절반(head)과 반전된 뒷 절반(pre)을 비교
        while pre:
            if pre.val != head.val:
                return False   # 값이 다르면 회문 아님
            pre = pre.next
            head = head.next

        # 4. 모두 통과했으면 회문
        return True
```
